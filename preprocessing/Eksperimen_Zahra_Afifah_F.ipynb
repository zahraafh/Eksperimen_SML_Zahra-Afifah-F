# **4. Exploratory Data Analysis (EDA)**

Pada tahap ini, Anda akan melakukan **Exploratory Data Analysis (EDA)** untuk memahami karakteristik dataset.

Tujuan dari EDA adalah untuk memperoleh wawasan awal yang mendalam mengenai data dan menentukan langkah selanjutnya dalam analisis atau pemodelan.

```python
df.info()
```

```python
df.isna().sum()
```

```python
df["Churn"].value_counts().plot(kind="bar")
plt.title("Distribusi Target Churn")
plt.xlabel("Churn")
plt.ylabel("Jumlah")
plt.show()
```

```python
df["Churn"].value_counts()
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"],errors="coerce")
```

```python
corr = df[["MonthlyCharges", "TotalCharges"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Korelasi Fitur Numerik")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df["MonthlyCharges"])
plt.title("Boxplot Monthly Charges")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df["TotalCharges"])
plt.title("Boxplot Total Charges")
plt.show()
```

Cukup sesuaikan dengan karakteristik data yang kamu gunakan yah. Khususnya ketika kami menggunakan data tidak terstruktur.

```python
df.isna().sum()
```

```python
#Mengisi missing value dengan median
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())
df["TotalCharges"].isna().sum()
```

```python
df.duplicated().sum() # tidak ada data duplikat
```

```python
# deteksi outlier menggunakan boxplot, dan sudah di lakukan di bagian EDA
```

```python
df.drop(columns=["customerID"], inplace=True)
```

```python
# one-hot encoding
categorical_cols = df.select_dtypes(include="object").columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
df.info()
```

```python
# Mengubah boolean ke int
bool_cols = df.select_dtypes(include="bool").columns
df[bool_cols] = df[bool_cols].astype(int)
df.info()
```
